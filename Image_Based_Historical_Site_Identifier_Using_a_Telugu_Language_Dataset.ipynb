```python
# üìå Step 1: Upload your image
from google.colab import files
uploaded = files.upload()

# Get uploaded image path
uploaded_image_path = list(uploaded.keys())[0]
print(f"‚úÖ Uploaded image: {uploaded_image_path}")

# üìå Step 2: Import required libraries
import cv2
import numpy as np
import pandas as pd
from skimage.metrics import structural_similarity as ssim
from PIL import Image
import matplotlib.pyplot as plt

# üìå Step 3: Define dataset in code (or you can later replace with Excel)
data = [
    {"‡∞ö‡∞ø‡∞§‡±ç‡∞∞ ‡∞´‡±à‡∞≤‡±Å ‡∞™‡±á‡∞∞‡±Å": "arunachalam.jpg", "‡∞∏‡±ç‡∞•‡∞≤‡∞Ç ‡∞™‡±á‡∞∞‡±Å": "‡∞Ö‡∞∞‡±Å‡∞£‡∞æ‡∞ö‡∞≤‡∞Ç", "‡∞µ‡∞ø‡∞µ‡∞∞‡∞£": "‡∞Ö‡∞∞‡±Å‡∞£‡∞æ‡∞ö‡∞≤‡±á‡∞∂‡±ç‡∞µ‡∞∞ ‡∞Ü‡∞≤‡∞Ø‡∞Ç ‡∞§‡∞Æ‡∞ø‡∞≥‡∞®‡∞æ‡∞°‡±Å‡∞≤‡±ã ‡∞â‡∞Ç‡∞¶‡∞ø. ‡∞á‡∞¶‡∞ø ‡∞∂‡±à‡∞µ ‡∞≠‡∞ï‡±ç‡∞§‡±Å‡∞≤‡∞ï‡±Å ‡∞™‡∞µ‡∞ø‡∞§‡±ç‡∞∞ ‡∞∏‡±ç‡∞•‡∞≤‡∞Ç‡∞ó‡∞æ ‡∞™‡∞∞‡∞ø‡∞ó‡∞£‡∞ø‡∞Ç‡∞ö‡∞¨‡∞°‡±Å‡∞§‡±Å‡∞Ç‡∞¶‡∞ø."},
    {"‡∞ö‡∞ø‡∞§‡±ç‡∞∞ ‡∞´‡±à‡∞≤‡±Å ‡∞™‡±á‡∞∞‡±Å": "jogulamba.jpg", "‡∞∏‡±ç‡∞•‡∞≤‡∞Ç ‡∞™‡±á‡∞∞‡±Å": "‡∞ú‡±ã‡∞ó‡±Å‡∞≤‡∞æ‡∞Ç‡∞¨", "‡∞µ‡∞ø‡∞µ‡∞∞‡∞£": "‡∞ú‡±ã‡∞ó‡±Å‡∞≤‡∞æ‡∞Ç‡∞¨ ‡∞Ü‡∞≤‡∞Ø‡∞Ç ‡∞Ü‡∞Ç‡∞ß‡±ç‡∞∞‡∞™‡±ç‡∞∞‡∞¶‡±á‡∞∂‡±ç ‡∞≤‡±ã‡∞®‡∞ø ‡∞Ö‡∞≤‡∞Ç‡∞™‡±Ç‡∞∞‡±ç ‡∞≤‡±ã ‡∞â‡∞Ç‡∞¶‡∞ø. ‡∞á‡∞¶‡∞ø ‡∞∂‡∞ï‡±ç‡∞§‡∞ø‡∞™‡±Ä‡∞†‡∞æ‡∞≤‡∞≤‡±ã ‡∞í‡∞ï‡∞ü‡∞ø."},
    {"‡∞ö‡∞ø‡∞§‡±ç‡∞∞ ‡∞´‡±à‡∞≤‡±Å ‡∞™‡±á‡∞∞‡±Å": "wayanand.jpg", "‡∞∏‡±ç‡∞•‡∞≤‡∞Ç ‡∞™‡±á‡∞∞‡±Å": "‡∞µ‡∞Ø‡∞®‡∞æ‡∞°‡±ç ‡∞ó‡±Å‡∞π‡∞≤‡±Å", "‡∞µ‡∞ø‡∞µ‡∞∞‡∞£": "‡∞à ‡∞ó‡±Å‡∞π‡∞≤‡±ç‡∞≤‡±ã BC 6000‡∞ï‡∞ø ‡∞ö‡±Ü‡∞Ç‡∞¶‡∞ø‡∞® ‡∞∂‡∞ø‡∞≤‡∞æ‡∞∂‡∞æ‡∞∏‡∞®‡∞æ‡∞≤‡±Å ‡∞â‡∞®‡±ç‡∞®‡∞æ‡∞Ø‡∞ø. ‡∞á‡∞¶‡∞ø ‡∞¶‡∞ï‡±ç‡∞∑‡∞ø‡∞£ ‡∞≠‡∞æ‡∞∞‡∞§‡∞¶‡±á‡∞∂‡∞™‡±Å ‡∞Æ‡±ä‡∞¶‡∞ü‡∞ø ‡∞Æ‡∞æ‡∞®‡∞µ ‡∞®‡∞ø‡∞µ‡∞æ‡∞∏‡∞æ‡∞≤‡∞ï‡±Å ‡∞ö‡∞ø‡∞π‡±ç‡∞®‡∞Ç‡∞ó‡∞æ ‡∞®‡∞ø‡∞≤‡±Å‡∞∏‡±ç‡∞§‡±Å‡∞Ç‡∞¶‡∞ø."}
]
df = pd.DataFrame(data)

# üìå Step 4: Load input image
input_img = cv2.imread(uploaded_image_path)
input_img = cv2.resize(input_img, (300, 300))
input_gray = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)

# üìå Step 5: Prepare reference images dictionary
reference_images = {}
for idx, row in df.iterrows():
    image_name = row["‡∞ö‡∞ø‡∞§‡±ç‡∞∞ ‡∞´‡±à‡∞≤‡±Å ‡∞™‡±á‡∞∞‡±Å"]
    try:
        img = cv2.imread(image_name)
        if img is not None:
            img = cv2.resize(img, (300, 300))
            reference_images[image_name] = img
    except:
        print(f"‚ö†Ô∏è Could not load: {image_name}")

# üìå Step 6: Compare input image with dataset images
best_score = -1
best_row = None
best_match_image = None

for idx, row in df.iterrows():
    image_name = row["‡∞ö‡∞ø‡∞§‡±ç‡∞∞ ‡∞´‡±à‡∞≤‡±Å ‡∞™‡±á‡∞∞‡±Å"]
    if image_name not in reference_images:
        continue

    ref_img = reference_images[image_name]
    ref_gray = cv2.cvtColor(ref_img, cv2.COLOR_BGR2GRAY)

    score, _ = ssim(input_gray, ref_gray, full=True)

    if score > best_score:
        best_score = score
        best_row = row
        best_match_image = ref_img

# üìå Step 7: Show results
print("\nüîç Best Match Result:")
if best_row is not None:
    for col in best_row.index:
        print(f"{col}: {best_row[col]}")
    print(f"\nüü¢ Image Similarity Score: {best_score:.2f}")

    # Show uploaded image and matched image side-by-side
    input_rgb = cv2.cvtColor(input_img, cv2.COLOR_BGR2RGB)
    match_rgb = cv2.cvtColor(best_match_image, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(input_rgb)
    plt.title("üì§ Uploaded Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(match_rgb)
    plt.title("‚úÖ Best Match from Dataset")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
else:
    print("‚ùå No matching image found in the dataset.")
```
