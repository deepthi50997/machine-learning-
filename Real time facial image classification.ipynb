```python
# ✅ STEP 1: Install necessary packages
!pip install -U gradio opencv-python keras tensorflow

# ✅ STEP 2: Download pretrained model
!wget https://github.com/oarriaga/face_classification/raw/master/trained_models/emotion_models/fer2013_mini_XCEPTION.102-0.66.hdf5 -O emotion_model.h5

# ✅ STEP 3: Import libraries
import cv2
import numpy as np
from keras.models import load_model
from keras.preprocessing.image import img_to_array
import gradio as gr
from PIL import Image
import os

# ✅ STEP 4: Load the model
model = load_model("emotion_model.h5", compile=False)

# ✅ STEP 5: Emotion labels
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

# ✅ STEP 6: Load face detector
# Construct the path to the Haar cascade file
cascade_path = os.path.join(cv2.data.haarcascades, "haarcascade_frontalface_default.xml")
face_detector = cv2.CascadeClassifier(cascade_path)

# ✅ STEP 7: Define prediction function
def detect_emotion(image: Image.Image) -> str:
    # Convert PIL to OpenCV
    image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect face using OpenCV's Haar cascade
    faces = face_detector.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    if len(faces) == 0:
        return "No face detected."

    for (x, y, w, h) in faces:
        roi = gray[y:y+h, x:x+w]
        roi = cv2.resize(roi, (64, 64))
        roi = roi.astype("float") / 255.0
        roi = img_to_array(roi)
        roi = np.expand_dims(roi, axis=0)

        preds = model.predict(roi)[0]
        emotion = emotion_labels[np.argmax(preds)]
        return f"Predicted Emotion: {emotion} {'😃' if emotion == 'Happy' else ''}"

    return "Face not clear enough."

# ✅ STEP 8: Gradio UI with webcam support
interface = gr.Interface(
    fn=detect_emotion,
    inputs=gr.Image(type="pil", label="📷 Live Webcam Input"),
    outputs=gr.Textbox(label="🙂 Predicted Emotion"),
    title="🧠 Real-Time Facial Expression Recognition",
    description="Use your webcam to capture your face and predict your current emotion in real-time using a CNN model.",
    theme="soft"
)

# ✅ STEP 9: Launch the interface
interface.launch(share=True)
```
