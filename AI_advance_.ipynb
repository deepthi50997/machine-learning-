```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[10,25,20,30])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line plot')
plt.grid(True)
plt.show()
```

```python
plt.bar(['A','B','C'],[5,10,20])
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('bar plot')
plt.show()
```

```python
plt.plot([1,2,3,4], [10,20,30,40], marker='o', linestyle='-', color='b', label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('cus line plot')
plt.legend()
plt.show()
```

```python
labels=['A','B','C']
sizes=[10,20,80]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(10,10))
axes[0,0].plot([1,2,3,4],[20,70,90,10],'r--')
axes[0,0].set_title('Subplot 1')

axes[0,1].bar(['A','B','C'],[5,10,16],color='g')
axes[0,1].set_title('Subplot 2')

axes[1,0].pie([10,20,80],labels=['A','Z','X'], autopct='%1.1f%%')
axes[1,0].set_title('Subplot 3')

fig.delaxes(axes[1,1])
plt.tight_layout()
plt.show()
```
