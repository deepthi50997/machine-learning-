```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn .model_selection import train_test_split
```

```python
#Breat cancer dataset
breast=sklearn.datasets.load_breast_cancer()
```

```python
print(breast)
```

```python
data_f=pd.DataFrame(breast.data,columns=breast.feature_names)
```

```python
data_f['label']=breast.target
```

```python
X=data_f.drop(columns='label',axis=1)
Y=data_f['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
print(X.shape,x_train,x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up layers
model=keras.Sequential([

                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')

])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['train','validation_data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['train','validation_data'],loc='upper right')
```

accuracy on test data

```python
loss,accuracy=model.evaluate(x_test_std,y_test)
print(accuracy)
```
